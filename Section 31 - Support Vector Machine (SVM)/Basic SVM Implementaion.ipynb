# Support Velctor Classifier Implementation

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## lets create synthetic datapoints

from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=2,n_redundant=0)
```

```python
x
```

```python
y
```

```python
pd.DataFrame(x)
```

```python
sns.scatterplot(x=pd.DataFrame(x)[0],y=pd.DataFrame(x)[1],hue=y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=.2,random_state=42)
```

```python
from sklearn.svm import SVC
```

# SVM kernel = Linear

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python

linear_svc=SVC(kernel="linear")
linear_svc.fit(x_train,y_train)

#Prediction
y_pred=linear_svc.predict(x_test)

print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
linear_svc.coef_
```

# SVM kernel = RBF

```python
rbf_svc=SVC(kernel="rbf")
rbf_svc.fit(x_train,y_train)

#Prediction
y_pred1=rbf_svc.predict(x_test)

print(classification_report(y_test,y_pred1))
print(confusion_matrix(y_test,y_pred1))
```

```python
rbf_svc.coef_
```

# SVM kernel = polynomial

```python
polynomial_svc=SVC(kernel="poly")
polynomial_svc.fit(x_train,y_train)

#Prediction
y_pred2=polynomial_svc.predict(x_test)

print(classification_report(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
```

# SVM kernel = sigmoid

```python
sigmoid_svc=SVC(kernel="sigmoid")
sigmoid_svc.fit(x_train,y_train)

#Prediction
y_pred3=sigmoid_svc.predict(x_test)

print(classification_report(y_test,y_pred3))
print(confusion_matrix(y_test,y_pred3))
```

```python
sigmoid_svc.intercept_
```

```python
# RBF kernel gives more accuracy 91%
```

## Hyperparametere Tuning With SVC

```python
from sklearn.model_selection import GridSearchCV
 
# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']}
```

```python
grid =GridSearchCV(SVC(),param_grid=param_grid,refit=True,cv=5,verbose=3)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python

#Prediction
y_pred4=grid.predict(x_test)

print(classification_report(y_test,y_pred4))
print(confusion_matrix(y_test,y_pred4))
```
